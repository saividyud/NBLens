```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import sys
sys.path.append('..')

from IRSMicroLensing import IRSMicroLensSimulation as IRSMLS
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
```

# Generating images

```python
pixels = 1000
ang_width = 5

source_att = [
    [0.2, 0.2, 0.1, 'Gauss']
]

lens_att = [
    [0, 0, 0.05, 1]
]

im = IRSMLS.IRSPlotter(pixels=pixels, ang_width=ang_width, source_att=source_att, lens_att=lens_att)
im.plot(save=True)

plt.show()
```

# Generating light curve

```python
pixels = 1000
ang_width = 5
rays_per_pixel = 4

r = np.sqrt(2.3)
theta = np.deg2rad(30)

lens_att = [
    [0, 0, 0.01, 1],
    [r*np.cos(theta), r*np.sin(theta), 0.01, 0.001]
]

caus = IRSMLS.IRSCaustics(pixels=pixels, ang_width=ang_width, lens_att=lens_att, rays_per_pixel=rays_per_pixel)
caus.plot(save_plot=False, show_lenses=True)

plt.show()
```

```python
pixels = 3000
ang_width = 5

r = np.sqrt(2.3)
theta = np.deg2rad(30)

lens_att = [
    [0, 0, 0.01, 1],
    [r*np.cos(theta), -r*np.sin(theta), 0.01, 0.001]
]

lc = IRSMLS.IRSLightCurves(pixels=pixels, ang_width=ang_width, lens_att=lens_att, source_r=0.05, u0=0.4, source_type='uniform', steps=300)
lc.plot(show_mm=False, save=True)

# fig = plt.figure(figsize=(10, 8))
# ax = fig.add_subplot()

# ax.plot(lc.time, lc.analytic_mags)

plt.show()
```
